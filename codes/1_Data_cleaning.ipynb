```python
dfnew['extract1'] =  dfnew['citations'].str.split('journal =').str[1].str.split('|').str[0]
dfnew['extract2'] =  dfnew['citations'].str.split('journal=').str[1].str.split('|').str[0]
dfnew['extract3']=dfnew['extract1'].combine_first(dfnew['extract2'])
#drop NaN
dfnew = dfnew.dropna(subset=['extract3'])
dfnew
```

```python
#Removing journal and magazine
dfnew['extract4'] =  dfnew['extract3'].str.replace('\(journal\)', '')
dfnew['extract4'] =  dfnew['extract4'].str.replace('\(magazine\)', '')
dfnew
```

```python
dfnew['extract4'] =  dfnew['extract4'].str.replace('\[\[', '')
dfnew['extract4'] =  dfnew['extract4'].str.replace('\]\]', '')
dfnew['extract4'] =  dfnew['extract4'].str.replace('\}\}', '')
```

```python
dfnew
```

```python
#Deleting white spaces

dfnew['extract4'] =  dfnew['extract4'].replace(r'[^\w\s]+', '', regex=True)
dfnew['extract4'] =  dfnew['extract4'].replace(r'[^\w\s]+', '', regex=True)
dfnew['extract4'] =  dfnew['extract4'].replace('The Journal of Biological Chemistry','Journal of Biological Chemistry')
dfnew['extract4'] =  dfnew['extract4'].replace('J Biol Chem','Journal of Biological Chemistry')
dfnew['extract4'] =  dfnew['extract4'].replace('Proc Natl Acad Sci USA','PNAS')
dfnew['extract4'] =  dfnew['extract4'].replace('Proceedings of the National Academy of Sciences','PNAS')
dfnew['extract4'] =  dfnew['extract4'].replace('Proceedings of the National Academy of Sciences of the United States of America','PNAS')
dfnew['extract4'] =  dfnew['extract4'].replace('PLOS ONE','PLoS ONE')
dfnew['extract4'] =  dfnew['extract4'].replace('PLOS One','PLoS ONE')
dfnew['extract4'] =  dfnew['extract4'].replace('Biochem Biophys Res Commun','Biochemical and Biophysical Research Communications')
dfnew['extract4'] =  dfnew['extract4'].replace('Biochem J','The Biochemical Journal')
dfnew['extract4'] =  dfnew['extract4'].replace('The Lancet','Lancet')
dfnew['extract4'] =  dfnew['extract4'].drop_duplicates()
dfnew['extract4'] =  dfnew['extract4'].str.strip()
dfnew
```

```python
df_final = dfnew.drop(['extract1', 'extract2', 'extract3'], axis = 1)
df_final = df_final.rename(columns = {'extract4':'journal_name'})
df_final
```

```python
#Deleting white spaces
df_final['journal_name'] = df_final['journal_name'].replace(r'[^\w\s]+', '', regex=True)


df_final['journal_name'] = df_final['journal_name'].str.lower()
df_final['journal_name'] = df_final['journal_name'].str.strip()
df_final['page_title'] = df_final['page_title'].str.lower()
df_final['page_title'] = df_final['page_title'].str.strip()
df_final = df_final.replace('the journal of biological chemistry','journal of biological chemistry')
df_final = df_final.replace('j biol chem','journal of biological chemistry')
df_final = df_final.replace('proc natl acad sci usa','pnas')
df_final = df_final.replace('proceedings of the national academy of sciences','pnas')
df_final = df_final.replace('proceedings of the national academy of sciences of the united states of america','pnas')
df_final = df_final.replace('biochem biophys res Commun','biochemical and biophysical research communications')
df_final = df_final.replace('biochem j','the biochemical journal')
df_final = df_final.replace('the lancet','lancet')
df_final = df_final.drop_duplicates()
df_final
```
